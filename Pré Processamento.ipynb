```python
# Fazendo a exportação da base de dados 

dfShopping = pd.read_csv('aluguel.csv', sep=";")
dfShopping
```

```python
#Separando apenas pelo Tipo utilizado para o trabalho, no caso Loja Shopping/ Ct Comercial

dfdataframe = dfShopping.loc[dfShopping["Tipo"] == "Loja Shopping/ Ct Comercial"]
dfdataframe
```

```python
#Ajustando Rótulos de linhas e colunas

linhas = list(range(247))
linhas
```

```python
#Ajustando Rótulos de linhas e colunas

dfdataframe.index = linhas
dfdataframe
```

```python
# ---------------------------- Pergunta 2: Criando seleções ------------------------------------------------------------

# 1° Seleção
# Selecionar Shoppings presentes no bairro "Barra da Tijuca"

bol = dfdataframe['Bairro'] == 'Barra da Tijuca'
bol
```

```python
# 2° Seleção
# Selecionar todos os Shopping que possuam Área maior que 50m2 e menor que 90m2 icluído

bol = (dfdataframe['Area'] >= 50) & (dfdataframe['Area'] <= 90)
dfdataframe[bol]
```

```python
#3° Seleção

# Selecionar todos os Shopping que possuam n° de quartos = 0 e vagas = 2

bol = (dfdataframe['Quartos'] == 0) & (dfdataframe['Vagas'] == 2)
dfdataframe[bol]
```

```python
#4° Seleção

# Selecionar todos os Shopping que possuam valor acima ou igual a 10000.0 ou que possuam área menor ou igual a 100

bol = ((dfdataframe['Valor'] >= 10000.0)) | ((dfdataframe['Area'] <= 100))
dfdataframe[bol]
```

```python
# 5° Seleção

# Selecionar todos os Shopping do Centro ou da Copacabana que tenham Valor de IPTU maior ou igual a 500

bol = ((dfdataframe['Bairro'] == "Centro") | (dfdataframe['Bairro'] == "Copacabana")) & (dfdataframe['IPTU'] >= 500)
dfdataframe[bol]
```

```python
# ---------------------------- Pergunta 3: Chame o método info() para o novo dataframe, para que seja possível -------------
# ---------------------------------------- verificar quais atributos possuem dados faltantes. -----------------------------
#----------------------------------------- Para cada atributo com dados faltante gere um dataframe para exibi-los --------


dfdataframe.info()
```

```python
# Pergunta 4: Proponha 2 possíveis transformações para o dataframe utilizando os métodos apply() e/ou map().

#Função para retornar "sem vagas" quando valor for = 0 e retornar "há vagas" para valor = 1

def QuartosVagas(qv):
    if qv == 0:
        return "Sem vagas"
    else:
        return "Há vagas"
```

```python
QuartosVagas(0)
```

```python
dfdataframe['Vagas'] = dfdataframe['Vagas'].apply(QuartosVagas)
dfdataframe
```

```python
# Dataframe para exibir os dados faltantes do atributo IPTU

dfdataframe[dfdataframe['IPTU'].isnull()]
```

```python
#Foi optado para esta base de dados preencher todos os valores ausentes com 0 pois cada imóvel possui 
# valores iptu de acordo com outros fatores (area, localidade, etc.)

dfdataframe = dfdataframe.fillna(0)
```

```python
dicionario = {'Barra da Tijuca':1, 'Botafogo':2, 'Centro':3, 'Campo Grande':4, 'Guadalupe':5, 'Gávea':6, 'Irajá':7, 'Jacarapaguá':8, 'Leblon':9, 'Ricardo de Albuquerque':10, 'Recreio dos Bandeirantes':11, 'Santa Cruz':12, 'São Cristovão':13, 'Vila da Penha':14}
```

```python
dfdataframe['Bairro'] = dfdataframe['Bairro'].map(dicionario)
dfdataframe
```

```python
dfdataframe.to_csv('minhabase.csv', index=False)
```
